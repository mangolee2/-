```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.metrics.pairwise import cosine_similarity

```

```python
# 1. 데이터셋 불러오기
df = pd.read_csv('C:/Users/seohe/OneDrive/바탕 화면/Thesis/dataset_user.csv')
```

```python
# data를 0으로 채우기
df=df.fillna(0)
```

```python
# 2. 특징 벡터 추출
X = df[['Activity', 'Location Frequency', 'Main Activity Time']].values
```

```python
# dtype을 'int128'로 설정하여 ValueError 회피
# 데이터 타입 변경
feature_vector = df.astype('float64') 
```

```python
# 3. 코사인 유사도 계산
cos_sim = cosine_similarity(X)
```

```python
# 4. FCM 알고리즘 적용
# KMeans 알고리즘 적용
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)
```

```python
# 5. 클러스터 결과 확인
labels = kmeans.labels_
```

```python
print(labels)
```

```python
# 클러스터링 결과 출력
for i in range(2):
    print(f"Cluster {i}: {df[labels == i]['User ID'].values}")
```

결과 시각하ㅗ

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
# t-SNE를 이용하여 2차원으로 축소
tsne = TSNE(n_components=2, random_state=0)
X_2d = tsne.fit_transform(X)
```

```python
# 클러스터링 결과 시각화
colors = ['red', 'blue', 'green', 'purple', 'orange', 'gray', 'black']
plt.figure(figsize=(10, 10))
for i in range(2):
    plt.scatter(X_2d[labels == i, 0], X_2d[labels == i, 1], color=colors[i], label=f"Cluster {i}")
plt.legend()
plt.show()
```

```python
# 클러스터링 결과 시각화
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, alpha=0.5)

# 클러스터 0에 해당되는 유저들의 id 출력
cluster_0_ids = df[labels == 0]['User ID'].values
print("Cluster 0 IDs:", cluster_0_ids)

# 클러스터 1에 해당되는 유저들의 id 출력
cluster_1_ids = df[labels == 1]['User ID'].values
print("Cluster 1 IDs:", cluster_1_ids)

plt.show()
```
