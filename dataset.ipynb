```python
df = pd.read_csv("C:/Users/seohe/OneDrive/바탕 화면/Thesis/data100.csv", encoding='cp949')
```

```python
# 오류 행 제거
df=df.dropna(axis=0)
```

```python
User=[]
for i in df['User ID']: 
    User.append(i)

print((User))
print(len(User))
```

```python
" 중복 제거한 유저 아이디 : UserTable1 "
UserTable = []

for i in User:
    if i not in UserTable:
        UserTable.append(i)
print(UserTable)
print(len(UserTable)) # 총 947 개의 유저id 
```

```python
data = pd.read_csv('C:/Users/seohe/OneDrive/바탕 화면/Thesis/dataset_df.csv')
```

```python
# print(data.columns)
```

```python
# 2. Time 열에서 시간 정보만 추출하기
# UTC time에서 시간 추출하기
data['Time'] = pd.to_datetime(data['UTC time'], format='%a %b %d %H:%M:%S +0000 %Y').dt.time
```

```python
# print((data['Time']))
```

```python
grouped_data = pd.read_csv('C:/Users/seohe/OneDrive/바탕 화면/Thesis/dataset_df_check.csv', header=None, names=['index', 'User ID' , 'Time', 'Main Activity Time'])
```

```python
# 3. 유저별로 그룹을 지어서 활동 시간대를 구하기
grouped_data = data.groupby('User ID')['Time'].apply(lambda x: pd.Series.mode(x)[0]).reset_index()
```

```python
# 4. 주된 활동 시간대를 설정하기
grouped_data['Main Activity Time'] = grouped_data['Time'].apply(lambda x: x.hour)
```

```python
grouped_data.to_csv('./dataset_df_check.csv')
```
