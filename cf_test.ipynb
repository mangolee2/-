```python
import pandas as pd

rating_data = pd.read_csv("C:/Users/seohe/OneDrive/바탕 화면/Thesis/testset/ratings.csv")

movie_data = pd.read_csv("C:/Users/seohe/OneDrive/바탕 화면/Thesis/testset/movies.csv")
```

```python
rating_data.head(2) # 사용자가 영화에 평점 매긴 데이터
```

```python
movie_data.head(2) # 영화 정보 
```

```python
rating_data.drop('timestamp', axis=1, inplace=True)
rating_data.head(2)
```

```python
user_movie_rating = pd.merge(rating_data, movie_data, on = 'movieId') # 겹치는 정보로 합침 
```

```python
user_movie_rating.head(2)
```

```python
movie_user_rating = user_movie_rating.pivot_table('rating', index = 'title', columns='userId') # 요걸로
user_movie_rating = user_movie_rating.pivot_table('rating', index = 'userId', columns='title')

# 우리는 지금 아이템 기반 협업 필터링할 예정이니, index 를 movie title로 둠, 당연히 column 정보는 userId 
```

```python
user_movie_rating.head(5)
```

```python
movie_user_rating.head() 
```

```python
movie_user_rating.fillna(0, inplace=True)
movie_user_rating.head(3)

# 평점을 매기지 않았으면 0, 매겼으면 1~5 점의 값 가지게 됨 
# 유사한 아이템 -> 평점이 비슷한 아이템 : 코사인 유사도 사용 
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_based_collabor = cosine_similarity(movie_user_rating)
item_based_collabor
```

```python
item_based_collabor = pd.DataFrame(data=item_based_collabor, index=movie_user_rating.index, columns=movie_user_rating.index)
```

```python
item_based_collabor.head()
```

```python
# 사용자가 그 영화가 마음에 들었다면, 그것과 비슷한 영화 추천

def get_item_based_collabor(title):
    return item_based_collabor[title].sort_values(ascending=False)[:6]
```
